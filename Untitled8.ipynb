```python
# Uji Hipotesis
import random
from scipy.stats import ttest_ind, ttest_rel
from scipy import stats

df = df[(np.abs(stats.zscore(df['Durasi']))) < 3]

income_1 = df[df['Credit_score'] == 1]['Durasi']
income_0 = df[df['Credit_score'] == 2]['Durasi']

income_0 = income_0.values.tolist()
income_0 = random.sample(income_0, 10)
income_1 = income_1.values.tolist()
income_1 = random.sample(income_1, 10)

ttest,pval = ttest_ind(income_1, income_0, equal_var = False)
print("ttest", ttest)
print("p-value", pval)

if pval < 0.05:
  print("Ho ditolak")
else:
  print("Ho diterima")
```

```python
def bar_chart(feature):
    Accepted = df[df['Credit_score']==1][feature].value_counts()
    Rejected = df[df['Credit_score']==2][feature].value_counts()
    df1 = pd.DataFrame([Accepted,Rejected])
    df1.index = ['Accepted','Rejected']
    df1.plot(kind='bar', stacked=True, figsize=(18, 6),title=feature)

bar_chart('Riwayat_kredit')
```

```python
# contingency
c_t = pd.crosstab(df['Riwayat_kredit'], df['Credit_score'], margins=False)
c_t
```

```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2

stat, p, dof, expected = chi2_contingency(c_t)
print('dof=%d' % dof)
print('p_value', p)
print(expected)

# interest test-statistic
prob = 0.99
critical = chi2.ppf(prob, dof)
print('probabilitas=%.3f, critical=%.3f, stat=%.3f' % (prob, critical, stat))
if abs(stat) >= critical:
  print('Dependen (tolak H0)')
else:
  print('Independen (terima H0)')
```
